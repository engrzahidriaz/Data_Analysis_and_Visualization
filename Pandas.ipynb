## nsmalest nlargest idxmin idxmax 

```python
guitar_dic = {'yamaha':8500,'faith apollo venus':30000,'Taylor':250000,'baton  Rounge':4000,'lakewood':30000}
```

```python
guitar_price = pd.Series(guitar_dic, name='Prices')
```

```python
guitar_price
```

```python
guitar_price.nsmallest() #give top 5 smallest guiter
```

```python
guitar_price.nsmallest(n=1) #give 1 value
```

```python
guitar_price.nsmallest().index[0]  #name of smallest guitar
```

```python
guitar_price.nlargest() #give top 5 largest guiter
```

```python
guitar_price.nlargest(n=2)
```

```python
guitar_price.nlargest().index[0]
```

```python
guitar_price.idxmin() #index value of smallest number
```

```python
guitar_price.idxmax() #index value of maximun number
```

```python
mask1 = df.country == 'Israel'
mask2 = df.year == 2014
df[mask1 & mask2]  # & use when both conditions are truth
```

```python
mask3 = df.year == 2015
f_df = df[mask1 & (mask2 | mask3)] 
```

```python
f_df
```

```python
f_df1 = df.loc[mask1 & (mask2 | mask3),['country','year','sex','age','suicides_no']] #give selected colunms
```

```python
f_df1
```

```python
mask4 = df.sex == 'male'
f_df2 = df.loc[mask1 & mask4 & (mask2 | mask3),['country','year','sex','age','suicides_no']] 
```

```python
f_df2
```

```python
f_df2.sum().suicides_no
```

```python
df.loc[df.year >=2010] #give data 2010 to onward
```

```python
mas1 = df.year >=2010
df.loc[mas1]
```

```python
mas2 = df.year.between(2000,2010) #data from 2000 to 2010
df.loc[mas2]
```

```python
mas3 = df.year.isin([2000,2010]) #just data of 2000 and 2010
df.loc[mas3]
```

```python
mas3 = df.year.isin([2000,2010]) 
df.loc[~mas3]    #~ sign use to not show data of 2000 and 2010, mean -ive
```

```python
mas3 = df.year.isin([2000,2010]) 
fil_df = df.loc[~mas3]
```

```python
fil_df.year.unique()
```

```python
# 2000 to 2014, male and female suicides_no of switzerland
mas4 = df.year.between(1990,2014)
mas5 = df.sex == 'male'
mas6 = df.country == 'Switzerland'
mas7 = df.age == '15-24 years'
filter_df = df.loc[mas4 & mas5 & mas6 & mas7, ['country','sex','year','suicides_no']]
```
