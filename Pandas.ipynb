# Pandas Tutorial

### Data Creation

```python
import pandas as pd

# pd.DataFrame(): Creates a DataFrame from scratch
data = {'title': ['Movie A', 'Movie B'], 'budget': [1000000, 2000000]}
example_df = pd.DataFrame(data)
example_df
```

```python
# Loading the CSV file into a DataFrame
movies_df = pd.read_csv("tmdb_5000_movies.csv")
movies_df
```

```python
# pd.Series(): Creates a Series from one column
popularity_series = pd.Series(movies_df["popularity"])
popularity_series
```

### Data Manipulation

```python
print("Displays the first 5 rows:\n", movies_df.head())
pd.DataFrame(movies_df.head())
```

```python
print("Displays the last 5 rows:\n", movies_df.tail())
pd.DataFrame(movies_df.tail())
```

```python
print("Summary of the DataFrame, column types, and non-null counts:\n", movies_df.info())
# datatype, range of index, total Columns (name, Non-Null Count,  Dtype), memory used  
```

```python
print("Provides descriptive statistics for numerical columns:\n", movies_df.describe())
pd.DataFrame(movies_df.describe())   # just show the integer and float data column
```

```python
avg_vote_by_language = movies_df.groupby("original_language")["vote_average"].mean()
print("Groups by 'original_language' and calculates the average 'vote_average':\n", avg_vote_by_language)
pd.DataFrame(movies_df.groupby("original_language")["vote_average"].mean())
```

```python
# DataFrame.merge(): Merges two DataFrames (Assuming `additional_df` contains movie data)
# merged_df = pd.merge(movies_df, additional_df, on="id")
```

```python
# pd.concat(): Concatenates DataFrames (Assuming two parts of the dataset are split)
# combined_df = pd.concat([movies_df_part1, movies_df_part2])
```

### Data Cleaning

```python
# DataFrame.isnull(): Detects missing values
print("Detects missing values:\n", movies_df.isnull().sum())
```

```python
# DataFrame.notnull(): Detects non-missing values
print("Detects non-missing values:\n", movies_df.notnull().sum())
```

```python
# DataFrame.dropna(): Removes rows with missing values
cleaned_df = movies_df.dropna()
cleaned_df
```

```python
# DataFrame.fillna(): Fills missing values with a default value
filled_df = movies_df.fillna("Unknown")
filled_df
```

```python
# DataFrame.replace(): Replaces specific values (e.g., replace all NaN in 'genres' column)
replaced_df = movies_df.replace({"genres": {"": "Unknown"}})
replaced_df
```

```python
# DataFrame.drop(): Deletes specified columns (e.g., 'homepage' column)
dropped_df = movies_df.drop(columns=["homepage"])
dropped_df
```

### Data Transformation

```python
# DataFrame.apply(): Applies a lambda function to create a new column based on existing data
movies_df["profit"] = movies_df.apply(lambda row: row["revenue"] - row["budget"], axis=1)
movies_df["profit"]
```

```python
# DataFrame.sort_values(): Sorts data by 'vote_average' in descending order
sorted_df = movies_df.sort_values("vote_average", ascending=False)
sorted_df
```

```python
# DataFrame.pivot_table(): Creates a pivot table to view average votes by language and status
pivot_table = movies_df.pivot_table(values="vote_average", index="original_language", columns="status", aggfunc="mean")
print(pivot_table)
```

```python
# DataFrame.agg(): Aggregates data, calculating multiple statistics for 'budget' and 'revenue'
aggregate_df = movies_df[["budget", "revenue"]].agg(["sum", "mean", "max", "min"])
print(aggregate_df)
```

### Analyzing and Visualizing Data

```python
# Top 10 most popular movies:
popular_movies = movies_df.nlargest(10, "popularity")[["title", "popularity"]]
popular_movies
```

```python
import matplotlib.pyplot as plt

# Plotting budget vs. revenue:
plt.scatter(movies_df["budget"], movies_df["revenue"])
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.title("Budget vs. Revenue")
plt.show()
```
